```python
import instaloader
import csv
from instaloader import Post

# # Membuat instance Instaloader
# L = instaloader.Instaloader()

# # Memuat sesi yang diimpor
# L.load_session_from_file('herq890')

# # Melanjutkan operasi lainnya
```

```python
link_content = input("masukan link komentar instagram? ")
link_content = link_content.split("p/")[1].split("/")[0]
nama = input("masukan nama file csv? ")
```

```python
# Membuat instance Instaloader
L = instaloader.Instaloader()

# Memuat sesi yang telah disimpan
L.load_session_from_file('herq890')

# Shortcode dari postingan
# SHORTCODE = 'DCWVLsbz7LO'
SHORTCODE = link_content

# Membuat objek Post
post = Post.from_shortcode(L.context, SHORTCODE)

# 
```

```python
# Mengambil komentar
comments = []
for comment in post.get_comments():
    comments.append({
        'id': comment.id,
        'username': comment.owner.username,
        'text': comment.text,
        'created_at': comment.created_at_utc
    })

# Menampilkan komentar
for comment in comments:
    print(f"{comment['username']} ({comment['created_at']}): {comment['text']}")


# save dataset
# Menyimpan komentar ke dalam file CSV
nama_file = f'komentar {nama}.csv'
with open(nama_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['id', 'username', 'text', 'created_at'])
    writer.writeheader()
    writer.writerows(comments)

print(f"Komentar telah disimpan ke dalam '{nama_file}'")
```
